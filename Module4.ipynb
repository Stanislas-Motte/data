### Associations of Variables

#### Covariance

We quantify association between two variables X and Y with Covariance

\begin{equation}
\mathrm{cov}(X,Y) = \frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})
\end{equation}

[Intuitively](https://www.youtube.com/watch?v=qtaqvPAeEJY), if both data points are below the mean the diffrences will both be of the same sign, therefore a positive term will be added, and vice versa. 

Covariance provides information about the relationship between two variables. It can tell can classify the relationship between two variabels into three categories: 
 
- (i) two variables tend to move together (+) 

- (ii) two variables move in the opposite directions (-)

- (iii) two variables do not have a relationship

**Covariance is only slightly interesting in and of itself** (the statistic is not standardized and it is not a measure of anything but direction of the relationship). It is a computational stepping stone to an interesting calculation like correlation. 

#### Correlation
Correlation is the standardized version of covariance, to make sure that the unit of measure is not affecting our covariance statistic.

\begin{equation}
\frac{\mathrm{cov}(X,Y)}{\sigma_{X}\sigma_{Y}}
\end{equation}

Correlation tells us not only the direction of the relationship, but also how strong that relationship is.

## SIMPLE LINEAR REGRESSION

We look to explain a dependent variabel using a single coefficient and OLS to find line of best fit. 

$y_i = \beta_0 + \beta_1 * x_i + \epsilon_i $

```python
import pandas as pd
import statsmodels as sm
import matplotlib.pyplot as plt
```

```python
#First, we explore simple estimation of SSE

b0 = 0.1
b1 = 1
housing['GuessResponse'] = b0 + b1*housing['RM']
housing['observederror'] = housing['MEDV'] - housing['GuessResponse']

# plot your estimated line together with the points
plt.figure(figsize=(10, 10))
plt.title('Sum of sqaured error is {}'.format((((housing['observederror'])**2)).sum()))
plt.scatter(housing['RM'], housing['MEDV'], color='g', label='Observed')
plt.plot(housing['RM'], housing['GuessResponse'], color='red', label='GuessResponse')
plt.legend()
#plt.xlim(housing['RM'].min()-2, housing['RM'].max()+2)
#plt.ylim(housing['MEDV'].min()-2, housing['MEDV'].max()+2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression

X = housing.loc[:, 'RM'].values.reshape(-1, 1)  # values converts it into a numpy array
Y = housing.loc[:, 'MEDV'].values.reshape(-1, 1)

#Run Linear Regression with scikit learn
regressor = LinearRegression()  # create object for the class
regressor.fit(X, Y)  # perform linear regression
Y_pred = regressor.predict(X)  # make predictions

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
# Calculate regression statistics manually
n = X.shape[0]  # number of samples
p = 1  # number of predictors
Y_pred = regressor.predict(X)  # predicted values
resid = Y - Y_pred  # residuals
SSE = np.sum(resid**2)  # sum of squared errors
MSE = SSE / (n - p - 1)  # mean squared error
RMSE = np.sqrt(MSE)  # root mean squared error
SSR = np.sum((Y_pred - np.mean(Y))**2)  # regression sum of squares
SST = SSE + SSR  # total sum of squares
R2 = SSR / SST  # R-squared
adj_R2 = 1 - (1 - R2) * (n - 1) / (n - p - 1)  # adjusted R-squared

# Print regression statistics
print('Regression line is y = {}x + {}'.format(regressor.coef_[0][0], regressor.intercept_[0]))

print(f'Sum of Squared Errors (SSE): {SSE:.2f}')
print(f'Mean Squared Error (MSE): {MSE:.2f}')
print(f'Root Mean Squared Error (RMSE): {RMSE:.2f}')
print(f'Regression Sum of Squares (SSR): {SSR:.2f}')
print(f'Total Sum of Squares (SST): {SST:.2f}')
print(f'R-squared: {R2:.2f}')
print(f'Adjusted R-squared: {adj_R2:.2f}')
```

```python
housing['BestResponse'] = regressor.intercept_[0] + regressor.coef_[0][0]*housing['RM']
housing['error'] = housing['MEDV'] - housing['BestResponse']
```
